#Hver er meðalaukning á púls hjá þeim sem hlupu eftir því hvort þeir stunda líkamsrækt í 0, 0-5, 5+ tíma á viku?

```python
hljop.assign(aukning = hljop.seinniPuls - hljop.fyrriPuls)
```

```python
bins = pd.IntervalIndex.from_tuples([(np.NINF,0), (0, 5), (5, np.Inf)])
aefingar = (
            hljop
            .assign(aefing = pd.cut(hljop.likamsraekt, bins = bins))
            .assign(aukning = hljop.seinniPuls - hljop.fyrriPuls)
            .groupby('aefing')
            .aukning
            .mean()
            )

aefingar
```

#Finnið fyrir hvert ár það land sem hefur hæstar lífslíkur í hverri heimsálfu

```python
grouped = df.groupby(['year', 'continent'])


# select
country_life = df[['year','continent','country','lifeExp']]

# index fyrir maxið
result = grouped['lifeExp'].idxmax()
result = country_life.iloc[result].set_index(['year','continent'])
result
```

#Finnið fyrir hvert land fyrsta árið þar sem lífslíkur aukast ekki um 1% á milli fimm ára. Skilið NaN ef ekkert slíkt ár er til.

```python
diff = df['year']
ok = df.loc[df.year < 2010]
ok
```

```python
grouped = df.groupby('country')

result = []

for n, g in grouped:

    # prósentubreyting
    diff = g['lifeExp'].diff(periods=5) / g['lifeExp'].shift(5)

    try:
        first_year = g.loc[diff < 0.01, 'year'].iloc[0]
        
    # ef ekkert slíkt ár er til
    except IndexError as e:
        first_year = np.nan
    # minus fimm til að fá hvenær óhamingjan byrjaði
    result.append((n, first_year - 5))

fimm_ar_af_ohamingju = pd.DataFrame(result, columns = ['country', 'year'])
fimm_ar_af_ohamingju
```

#Finnið meðalfermetraverð á höfuðborgarsvæðinu fyrir fjölbýli á bilinu 60-120m2

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv("https://objectstorage.eu-frankfurt-1.oraclecloud.com/n/frs3o1zldvgn/b/public_data_for_download/o/kaupskra.csv", low_memory= False)

df['tegund'].unique()
```

```python
# taka í burtu þetta fáránlega whitespace
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df
```

```python
# höfuðborgarsvæðið

hbs = [ 276,271,270,225,222,221,220,212,210,206,203,
       202,201,200,172,170,162,161,155,150,132,130,
       129,128,127,125,124,123,121,116,113,112,111,
       110,109,108,107,105,104,103,102,101
    ]

```

```python
# Skoða gögnin
df[df.heimilisfang == 'Grenimelur 1']
```

```python
# eignir á höfuðborgarsvæðinu
hbs = df[df['postnr'].isin(hbs)]
```

```python
# fjölbýli
hbs_fjolbyli = hbs[hbs['tegund'] == 'Fjölbýli']
```

```python
# á milli 60 og 120 fmetrar
hbs_fjolbyli_60_120 = hbs_fjolbyli[hbs_fjolbyli['einflm'].between(60,120)]
```

```python
svar = (hbs_fjolbyli_60_120.
        assign(verd_per_fermetra = hbs_fjolbyli_60_120.kaupverd / hbs_fjolbyli_60_120.einflm)
        ['verd_per_fermetra'].
        mean()
        )
svar
```
